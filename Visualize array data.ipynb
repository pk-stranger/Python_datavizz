```python
import pandas as pd
```

```python
irisdata = pd.read_csv('D:\\basicdata\\iris.data',header =None)
irisdata
```

```python
import numpy as np
```

```python
data_np = irisdata.iloc[:,:-1].to_numpy()  #พิจารณาในรูปเเบบ array เอาตั้งเเต่หลักเเรกจนถึงหลักรองสุดท้าย [:,:-1] มันจะตัดตัวสุดท้ายออก //เเรกเเถว หลัง คือหลัก
data_np #เราอยากได้เเค่ตัวเลข
```

```python
#จะตัดข้อมูลเเค่ setosa
setosa_np = irisdata.iloc[:50,:-1].to_numpy()  
setosa_np
```

```python
versicolor_np = irisdata.iloc[50:101,:-1].to_numpy()  
versicolor_np
```

```python
virginica_np = irisdata.iloc[101:,:-1].to_numpy()  
virginica_np
```

## visualize array

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
plt.imshow(setosa_np,interpolation='nearest')
plt.show()
```

```python
#วิธีขยาย ให้มองเห็นได้ชัดด้วยการ เซ้ดไซต์
matplotlib.rcParams['figure.figsize'] = [12,12] #อยากให้ได้กว้างกับด้านสูงมีขนาดกี่นิ้ว
plt.imshow(setosa_np,interpolation='nearest') 
plt.show() #สีสว่างค่ามาก สีเข้มค่าน้อย
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12] #อยากให้ได้กว้างกับด้านสูงมีขนาดกี่นิ้ว
plt.imshow(versicolor_np,interpolation='nearest') 
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12] #อยากให้ได้กว้างกับด้านสูงมีขนาดกี่นิ้ว
plt.imshow(virginica_np,interpolation='nearest') 
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] #subplot กี่เเถว 1 เเถว 3 หลัก 1 รูปทึ่1 ,รูปที่2 ,3
plt.subplot(1,3,1)
plt.imshow(setosa_np,interpolation='nearest') 
plt.subplot(1,3,2)
plt.imshow(versicolor_np,interpolation='nearest') 
plt.subplot(1,3,3)
plt.imshow(virginica_np,interpolation='nearest') 
plt.show()
```

```python
#เปลี่ยน data ให้ plot เเยกชั้นกัน  คือplotทั้งตาราง
matplotlib.rcParams['figure.figsize']
plt.imshow(data_np, interpolation = 'nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12] 
plt.imshow(setosa_np, interpolation = 'nearest', cmap = plt.get_cmap('Greys'))  #ไล่สี ด้วยคำสั่ง cmap
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] #subplot กี่เเถว 1 เเถว 3 หลัก 1 รูปทึ่1 ,รูปที่2 ,3
plt.subplot(1,3,1)
plt.imshow(setosa_np,interpolation='nearest', cmap = plt.get_cmap('PRGn')) 
plt.subplot(1,3,2)
plt.imshow(versicolor_np,interpolation='nearest', cmap = plt.get_cmap('PRGn')) 
plt.subplot(1,3,3)
plt.imshow(virginica_np,interpolation='nearest', cmap = plt.get_cmap('PRGn')) 
plt.show()
```

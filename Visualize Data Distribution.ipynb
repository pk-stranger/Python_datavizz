```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=10*setosa_np[:,2], c=colors) 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 30*setosa_np[:,2], c = setosa_np[:,3])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 40*setosa_np[:,2], c = setosa_np[:,3], cmap = 'YlOrBr')
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=30*setosa_np[:,2], c=np.array([[1,0.5,0.8]]))  #กำหนดสีตั้งเเต่เลข0-1
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=30*setosa_np[:,2], c=setosa_np[:,3], cmap='YlOrBr',marker = "h") 
```

### Visualize 2 Data

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
versicolor_np
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=30*setosa_np[:,2], c=setosa_np[:,3], cmap='BuPu',marker = "h") 
plt.scatter(versicolor_np[:,0], versicolor_np[:,1], s=30*versicolor_np[:,2], c=versicolor_np[:,3], cmap='BuPu',marker = "*") 
```

## Visualize 3 Data

```python
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=30*setosa_np[:,2], c=setosa_np[:,3], cmap='BuPu',marker = "h") 
plt.scatter(versicolor_np[:,0], versicolor_np[:,1], s=30*versicolor_np[:,2], c=versicolor_np[:,3], cmap='BuPu',marker = "*") 
plt.scatter(virginica_np[:,0], virginica_np[:,1], s=30*virginica_np[:,2], c=virginica_np[:,3], cmap='BuPu',marker = "x") 
```

```python
plt.scatter(setosa_np[:,2], setosa_np[:,3], s=30*setosa_np[:,0], c=setosa_np[:,1], cmap='BuPu',marker = "h") 
plt.scatter(versicolor_np[:,2], versicolor_np[:,3], s=30*versicolor_np[:,0], c=versicolor_np[:,1], cmap='BuPu',marker = "*") 
plt.scatter(virginica_np[:,2], virginica_np[:,3], s=30*virginica_np[:,0], c=virginica_np[:,1], cmap='BuPu',marker = "x") 
```

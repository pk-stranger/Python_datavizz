## เชื่อมตาราง 2 ตารางเข้าด้วยกัน หรือเรียกว่าการรวมตาราง

### 1. ตัด column อีกตารางนึงมาแปะในอีกตารางนึง

#### ทำการเพิ่ม column ลงในตาราง small_table เพื่อให้ทั้งสองตารางเชื่อมกันได้ โดยเอา column ที่ต้องการมาจากตาราง w_restaurant_df

```python
small_table['id'] = w_restaurant_df['id']
```

```python
small_table.head()
```

### 2. .map() เป็นอีกหนึ่งคำสั่งที่ไว้เชื่อมตารางเช่นกัน

#### เพิ่ม business_name ซึ่งเป็นชื่อ column ใหม่ที่ต้องการจะสร้าง ลงไปใน search_click_df เพื่อเป็นการเพิ่มชื่อร้านอาหารลงไป
#### ตาราง search_click_df เป็นตารางหลัก column business_id map กับ ตาราง small_table ที่เป็นตารางรอง

```python
search_click_df.head()                  #ตารางเดิม
```

```python
search_click_df['business_name'] = search_click_df['business_id'].map(small_table.set_index('id')['name'])       #ตารางใหม่
search_click_df.head()
```

## .merge() คือคำสั่งรวมกัน

### รวม 2 ตารางเข้าด้วยกัน

![join_table](aaa.png)

#### ตั้งชื่อตารางใหม่ = ตารางหลัก ตามด้วย .merge ตารางรอง

```python
merged_table = search_click_df.merge(small_table, how = 'left', left_on = 'business_id', right_on = 'id')
merged_table.head()
```
